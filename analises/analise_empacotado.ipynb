```python
import pandas as pd
import plotly.express as px
```

```python
dados_empacotado = pd.read_excel('../Empacotado.xlsx')
dados_produzido = pd.read_csv('../dados_ordem.csv')
```

```python
dados_empacotado = dados_empacotado.rename(columns={'Corrida':'Ordem de prod'})
```

```python
dados_empacotado.head()
```

```python
dados_produzido.head()
```

### Agrupando total de empacotados por ordem

```python
empacotado_por_ordem = dados_empacotado.groupby('Ordem de prod')['Qtd Empacotado'].sum()
```

### Combinando dataframes

```python
df_producao = dados_produzido.merge(empacotado_por_ordem, on='Ordem de prod')
```

```python
df_producao
```

```python
df_producao = df_producao.rename(columns={'Total de Defeitos': 'Defeitos Inspeção Eventual', 'Qtd Empacotado': 'Total Empacotado'})
```

```python
df_producao['Total Não Empacotado'] = df_producao['Total de Frascos'] - df_producao['Total Empacotado'] 
```

```python
# Criando a nova ordem das colunas
nova_ordem = [
    'Ordem de prod', 'Ref. do Artigo', 'Duração da ordem', 'Temperatura (F°)', 
    'Peso médio (g)', 'Comprimento médio (mm)', 'Diâmetro (mm)', 'Área da gota (mm)', 
    'Gotas cortadas (toneladas)', 'Total de Frascos', 'Total Empacotado', 'Total Não Empacotado', 
    'Defeitos Inspeção Eventual', 'Terminação deformada', 'Racho no anel', 'Sujo de óleo', 'Cascão', 
    'Frasco caído na esteira', 'Dupla costura', 'Frio', 'Lascado em produção', 'Dobra', 'Trinca por choque térmico', 
    'Racho na costura', 'Racho na terminação', 'Cava no ombro', 'Ombro mal cheio', 
    'Ombro deformado', 'Martelado', 'Aderido'
]

# Reordenando as colunas do dataframe
df_producao = df_producao[nova_ordem]
```
